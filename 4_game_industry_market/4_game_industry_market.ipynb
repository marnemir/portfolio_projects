Сначала проверим годы на адекватность. Речь идёт о прогнозе на **2017 год**, т. е. год **не будет больше 2016**.  
Также проверим, чтобы не было нереалистичных **минимальных годов**. Например, нереалистичным можно считать год до 1970.

Теперь перейдем непосредственно к **пропускам**. Причиной пропуска *year_of_release* может быть:
- **Пролемы с запуском**: официальная дата одна, а по факту в продажу игра поступила в другое время.
- Игра была выпущена **в разное время для разных стран**. Например, официальный релиз был в конце года, но какие-то страны увидели игру только в начале следующего года, из-за чего нельзя однозначно назвать год.  

- **Отсутствие данных**. Под "отсутствием" можно подразумевать как отсутствие **официальных данных** (например, игра не очень значима и не очень популярна, поэтому по ней нет информации), так и просто отсутствие данных конкретно **в нашей базе**.


В случае пропуска по какой-либо платформе, можно посмотреть, есть ли игра на **других платформах** и взять тот же год выпуска.  
Проверим несколько значений:

```python
# проверка игры fifa soccer 2004
games.query('name_low == "fifa soccer 2004"')
```

На остальных платформах **игра вышла в 2003 году**.

```python
# проверка lego batman: the videogame
games.query('name_low == "lego batman: the videogame"')
```

Пропуски есть по двум платформам, однако для остальных платформ официальная дата выпуска **2008 год**.

Составим **список игр**, для которых есть пропуски в **годе выпуска**:

```python
game_null_list = games[games['year_of_release'].isnull()]['name_low'].unique()

print('Количество игр с пропусками:', len(game_null_list))
print()
print(game_null_list[0:20])
```

В примере с "Need for Speed: Most Wanted" мы видели, что у одной игры может быть **две даты выпуска** из-за перезапуска игры.  
В нашем случае будем придерживаться следующей логики: если игра выходила на **других платформах** и на других платформах **только один год выпуска**, то вместо пропуска укажем **год для остальных платформ**.  

```python
# количество пропусков ДО
print('Количество пропусков ДО:', games.year_of_release.isnull().sum())

# для игр из списка
for game in game_null_list:
    
    # срез с для игры
    game_series = games[games['name_low'] == game]
    
    # количество уникальных годов
    game_series_year_number = game_series['year_of_release'].nunique()
    
    # проверка что уникальный год один
    if game_series_year_number == 1:
    
        # если год один, заполняем медианой
        game_year_fill = game_series['year_of_release'].median()
        
        # заполнение пропуска медианой
        games.loc[(games['name_low'] == game) & (games['year_of_release'].isnull()), 'year_of_release'] = game_year_fill

# количество пропусков ПОСЛЕ        
print('Количество пропусков ПОСЛЕ:', games.year_of_release.isnull().sum())
```

Ушло **111 пропусков** - значит, для игры был только **один год выпуска** и она выпускалась в т. ч. **на других платформах**.

Здесь же можно допустить, что рейтинги игры примерно одинаковые и в целом **не противоречат друг другу**.  

Противоречием будет, если **рейтинг игры M (для взрослых)**, но при этом на одной из приставок будет рейтинг, например, **E10+**. Проверим, есть ли такие игры с противоречивыми рейтингами:

```python
# сделаем копию рабочего датафрейма
games_test_rate = games.copy()

# заменим пропуски на unknown
games_test_rate['rating'] = games_test_rate['rating'].fillna('unknown')

# создадим список с играми, к-е содержат рейтинг M и AO
test_list = games_test_rate.query('rating == "M" or rating == "AO"')['name_low'].unique()

# цикл для проверки
for game in test_list:
    gslice = games_test_rate.query("name_low == @game") # срез для игры
    grating_list = gslice['rating'].unique() # список уникальных рейтингов
    grating_list_fin = [i_game for i_game in grating_list if i_game!='unknown'] # если рейтинг неизвестен, убираем из списка
    
    # если после ухода пропусков осталось всё равно больше 1 рейтинга, вывести список
    if len(grating_list_fin) > 1: 
        print(game, end=' ')
        print(grating_list)
```

Каких-то явных противоречий типа **M** и **E** не встретилось, поэтому оставим рейтинги в текущем виду. 

Теперь обратимся к **заполнению пропусков**: если для игры **есть рейтинг на других платформах**, мы можем **заполнить этим рейтингом пропуск**. Задача осложняется тем, что у некоторых игр может быть **несколько рейтингов**. Но мы проверили, что каких-то явных противоречивых комбинаций типа M и E не встречалось, поэтому можем взять какой-либо один из них.

```python
print('Количество уникальных игр без рейтинга:', end=' ')
print(games[games['rating'].isnull()]['name_low'].nunique())
print(games[games['rating'].isnull()]['name_low'].unique())
```

```python
no_rating = games[games['rating'].isnull()]['name_low'].unique()
yes_rating = games[games['rating'].notnull()]['name_low'].unique()

fill_rating = []

for name in no_rating:
    if name in yes_rating:
        fill_rating.append(name)

        
print('Можно заполнить пропусков по играм:', len(fill_rating))
print(fill_rating[0:20])
```

Итак, можно заполнить пропуски **по 359 играм**. Будем пользоваться примерно той же логикой, что и для заполнения информации по *year_of_release*: создадим цикл, в котором пропуск будем заменяться каким-нибудь из имеющихся рейтингов.

```python
print('Количество пропусков ДО:', games.rating.isnull().sum())

# для игр из списка, где можно заполнить пропуски
for game in fill_rating:
    
    # срез с для игры, не учитывая пропуски в рейтингах
    game_series = games[(games['name_low'] == game) & (games['rating'].notnull())]
    
    # список уникальных рейтингов
    game_rating_list = game_series['rating'].unique()
    
    # берем первый рейтинг из списка
    game_rate_fill = game_rating_list[0]
    
    # заполнение пропуска с помощью game_rate_fill
    games.loc[(games['name_low'] == game) & (games['rating'].isnull()), 'rating'] = game_rate_fill
    
print('Количество пропусков ПОСЛЕ:', games.rating.isnull().sum())
```

Помимо пропусков, достаточно много оценок ***tbd***. В отношении медиаигр ***tbd (to be determined)*** означает, что у игры **недостаточно отзывов, чтобы сформировать пользовательский рейтинг**. В частности, у [Metacritic](https://www.metacritic.com/faq#item13) ***tbd*** означает, что игра получила **менее 4 отзывов**.

В нашем случае мы можем расценивать это значение как **пропуск** - оценки нет, она неизвестна.  
Заменим все значения ***tbd*** на **пропуск**, а также изменим тип данных колонки на *float*:


```python
print('Количество пропусков ДО:', games.user_score.isnull().sum())

# замена tbd на пропуск
games['user_score'] = games['user_score'].replace("tbd", np.nan).astype('float')

# проверка замены
print('Количество пропусков ПОСЛЕ:', games.user_score.isnull().sum())

# вывод уникальных значений
print(games.user_score.value_counts(dropna=False, normalize=True).head())
```

- **Масимальный** равен **9.7 баллам**.

### 3.5. Взаимосвязь продаж и отзывов пользователей и критиков

На примере актуальной популярной платформы рассмотрим взаимосвязь **продаж** *total_sales* и **отзывов критиков** *critic_score* и **пользователей** *user_score*. В качестве актуальной платформы выберем **PS4**.

```python
# срез для PS4
ps4_score = games.query('platform == "PS4"')

# расчет коэффициента корреляции
ps4_score[['user_score', 'critic_score', 'total_sales']].corr()
```

```python
# визуализация матрицы с корреляциями
pd.plotting.scatter_matrix(ps4_score[['user_score', 'critic_score', 'total_sales']], figsize=(7, 7))
plt.suptitle('Визуализация матрицы корреляций, платформа PS4: \n отзыв пользователей (user_score), отзыв критиков (critic_score), глобальные продажи (total_sales)')
plt.show()
```

- Наибольшее абсолютное значение коэффициента корреляции для **оценок пользователей** и **оценок критиков** (значение **0.55**). Это говорит о наличии **положительной взаимосвязи**. Если положительные оценки у критиков, то, скорее всего, положительные оценки и у пользователей, и наоборот.  

- Также **положительную взаимосвязь** можно проследить между **глобальными продажами** и **оценками критиков** - значение коэффициента **около 0.4**.  

- При этом между **оценками пользователей** и **продажами** коэффициент корреляции **отрицательный** и в принципе **близок к нулю**. На графиках видно, что это произошло из-за **выбросов** - по некоторым играм с крупными продажами есть несколько **низких** оценок, что влияет на общий результат.

Для большей наглядности построим отдельные диаграммы рассеяния для **critic_score и total_sales** и **user_score и total_sales**:

```python
def corr_scatter(df_name, col_name_x, col_name_y='total_sales',
                 xlabel=None, ylabel='Глобальные продажи (млн долларов)',
                 xticks=None, yticks=None):
    
    """Функция выводит коэффициент корреляции и диаграмму рассеяния:
    - df_name - наименование датафрейма;
    - col_name_x - наименовании колонки для оси X;
    - col_name_y (по умолчанию 'total_sales')- наименование колонки для оси Y;
    - xlabel - значение xlabel;
    - ylabel (по умолчанию 'Глобальные продажи (млн долларов)') - значение ylabel;
    - xticks - значение xticks;
    - yticks - значение yticks."""
    
    # вывод коэффициента корреляции
    print(f'Коэффициент корреляции для {col_name_x} и {col_name_y}:', end='')
    print(' {:.2%}'.format(df_name[col_name_x].corr(df_name[col_name_y])))
    
    # вывод диаграммы рассеяния
    df_name.plot(x=col_name_x, y=col_name_y, marker='.', linestyle='none', alpha=0.5, grid=True)
    plt.title(f'Диаграмма рассеяния для {col_name_x} и {col_name_y}')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    plt.xticks(xticks)
    plt.yticks(yticks)
    plt.legend(loc='upper left')

    plt.show()
```

```python
corr_scatter(ps4_score, 'critic_score', 
             xlabel='Количество баллов (0 - минимум, 100 - максимум)',
             xticks=np.arange(0, 101, 10), yticks=np.arange(15))
```

- Есть достаточно игр, по которым **продажи** были **не самыми выдающимся** (до 1 млн долларов), но при этом **оценки** были **разнообразными** (как высокие, так и низкие).
- При этом в данных **нет такой ситуации**, что игра была **дорогой** (глобальные продажи более 2 млн), а **оценки** у неё **низкие** (менее 60 баллов). 

Если отсечь **самые низкие оценки (менее 50 баллов)**, то показатель корреляции должен улучшиться. Проверим это:

```python
# ограничение для среза
score_limit_critic = ps4_score.query('critic_score > 50')

corr_scatter(score_limit_critic, 'critic_score', 
             xlabel='Количество баллов (0 - минимум, 100 - максимум)',
             xticks=np.arange(50, 101, 10), yticks=np.arange(15))
```

- Да, показатель стал немного лучше и поднялся с **40%** и **44%**. Тем не менее, смысл прежний: между оценками критиков и глобальными продажами можно проследить **положительную взаимосвязь**. 
- Здесь мы только констатируем факт, что **взаимосвязь есть**, но **не устанавливаем причинно-следственную связь**, т. е. нельзя однозначно сказать, что из-за положительных обзоров критиков повышаются глобальные продажи игры.

Проделаем аналогичную операцию для **отзывов пользователей**:

```python
corr_scatter(ps4_score, 'user_score', 
             xlabel='Количество баллов (0 - минимум, 10 - максимум)',
             xticks=np.arange(11), yticks=np.arange(15))
```

- Как было отмечено ранее, точечные **низкие оценки** для игр **с крупными продажами** влияют на всю картину в целом, из-за чего получается **отрицательная взаимосвязь**.
- Если **убрать эти выбросы**, то сильно коэффициент корреляции не улучшится, но, возможно, перестанет быть хотя бы отрицательным.
Проверим это:

```python
# ограничение для среза
score_limit_user = ps4_score[~((ps4_score['total_sales'] > 3) & (ps4_score['user_score'] < 6))]

corr_scatter(score_limit_user, 'user_score', 
             xlabel='Количество баллов (0 - минимум, 10 - максимум)',
             xticks=np.arange(11), yticks=np.arange(15))
```

- Значение коэффициента корреляции по-прежнему небольшое, но теперь оно **положительное**.

Рассмотрим подробнее ситуации, когда оценки пользователей **низкие**, но при это продажи достаточно **высокие**.  
Выведем список нескольких подобных игр:

```python
# глобальный бюджет более 5 млн, рейтинг пользователей менее 6
ps4_score.query('total_sales > 5 and user_score < 6')
```

По какой-то причине отзывы **критиков** достаточно благосклонные, чего нельзя сказать об отзывах **пользователей**.  
Посмотрим, какая ситуация происходила на других платформах. Для примера возьмём FIFA:

```python
# посмотрим на другие платформы и оценки критиков
games[(games['name_low'].isin(["fifa 15", "fifa 16", "fifa 17"])) & (games['critic_score'].notnull())].sort_values(by='name')
```

- Любопытно, что и **на других** доступных **платформах** оценки **критиков** достаточно **высокие**, а оценки пользователей средние.
- Возможно, в каких-то оценках есть **ошибка** или же в оценках критиков есть **предвзятость** (например, на самом деле отзывы заказные).

Это вопрос отдельного исследования. Здесь мы остановимся на том, что среди пользователей бывают ситуации, что **игры** с **высокими продажами** оцениваются достаточно **низко**.

Посмотрим, как соотносятся выводы с другими популярными платформами:

```python
def corr_scatter_matrix(df_name, platform_name):
    """Функция возвращает матрицу корреляций и визуализацию матрицы корреляций:
    - df_name - наименование датафрейма;
    - platform_name - наименование платформы."""
    # срез для платформы
    platform_slice = df_name[df_name['platform'] == platform_name]
    
    # расчет коэффициента корреляции
    print(platform_slice[['user_score', 'critic_score', 'total_sales']].corr())
    
    # визуализация матрицы с корреляциями
    pd.plotting.scatter_matrix(platform_slice[['user_score', 'critic_score', 'total_sales']], figsize=(7, 7))
    plt.suptitle(f'Визуализация матрицы корреляций, платформа {platform_name}: \n отзыв пользователей (user_score), отзыв критиков (critic_score), глобальные продажи (total_sales)')
    plt.show()
```

```python
# срез для XOne
corr_scatter_matrix(games, "XOne")
```

Ситуация с отзывами **XOne** похожа на ситуацию с **PS4**:

Мы рассмотрели наиболее и наименее прибыльные жанры. Но существует ещё один нюанс: мы выяснили, что высокие продажи свойственны **серийным играм**. Выведем топ по каждому жанру:

```python
def top20_games(df_name, genre_name):
    
    """Функция выводит топ-20 игр для выбранного жанра:
    - df_name - наименование датафрейма;
    - genre_name - наименование жанра."""
    
    result = (df_name[df_name['genre'] == genre_name]
                 .groupby('name')[['year_of_release', 'total_sales']]
                 .agg({'year_of_release': 'last', 'total_sales': 'sum'})
                 .sort_values(by='total_sales', ascending=False)
                 .head(20)
             )
    return result
```

```python
# топ для shooter
top20_games(games_final_act, 'shooter')
```

- В основном в топе находятся игры, у которых **много серий**.
- Тем не менее в топе встречаются и **свежие релизы** (например, "Titanfall", "Overwatch"\*). Есть шанс, что новый шутер с неизвестным именем найдет свою аудиторию и принесет хорошие продажи.

\* *На момент рассмотрения данных у этих игр не было сиквелов. Сейчас, например, у Titanfall появилась уже 2 часть.*

```python
# топ для sports
top20_games(games_final_act, 'sports')
```

- В играх жанра **sports** подавляющее количество **игр одного бренда**: каждый год для культовых игр тип "FIFA" выпускается очередная серия с новым сезоном, составом игроков и т. п. Бренды имеют своих преданных фанатов из-за чего у **одной и той же серии** высокие продажи.
- На первый взгляд запускать новый неизвестный бренд может быть достаточно рискованно.

```python
# топ для role-playing
top20_games(games_final_act, 'role-playing')
```

- Поэтому игра такого жанра может вызвать интерес и найти своего покупателя.

```python
# топ для platform
top20_games(games_final_act, 'platform')
```

- Видна всё та же **серийность**, но встречаются и **новые** для того времени названия, например, "Knack", "Tearaway Unfolded".
- У данного жанра нет таких же крупных хитов по продажам, как у жанров выше, но всё-таки он может принести неплохие продажи.

Напишем функцию, которая будет выводить графики *barchart* (суммарные продажи по региону) и *piechart* (доля продаж для топ-5):

```python
def top_sales_barh_pie(df_name, col_groupby, col_val, region_name, group_name, 
                       xlabel='Продажи (млн долларов)', ylabel=None, xticks=None):
    
    """Функция создаёт barchart (суммарные продажи по региону) и piechart (доля продаж для топ-5):
    - df_name - наименование датафрейма;
    - col_groupby - наименование колонки для группировки;
    - col_val - продажи региона;
    - region_name - наименование региона (в именительном падеже);
    - group_name - наименование группирующего признака (в именительном падеже);
    - xlabel (по умолчанию 'Продажи (млн долларов)') - значение xlabel;
    - ylabel - значение ylabel;
    - xticks - значение xticks."""
    
    # группировка продаж
    games_group = (df_name
                    .groupby(col_groupby)[col_val].sum()
                    .sort_values()
                   )
    print(f'Тотальные продажи ({region_name}):', end=' ')
    print('{:.2f}'.format(games_group.sum()))

    # построение barchart
    games_group.plot(kind='barh', grid=True, figsize=(8, 5))
    y = [np.round(i, 2) for i in games_group.values]
    for index, value in enumerate(y):
        plt.text(value, index, str(value))
    plt.title(f'Суммарные продажи, {group_name} ({region_name})')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(xticks)
    plt.show()
    
    # подготовка данных для piechart
    games_group = games_group.sort_values(ascending=False).reset_index()
    top5_pie = games_group.head()
    others = games_group.iloc[5:].sum().values[-1]
    top5_pie.loc[5] = ['Others', others]

    # значения для piechart
    plt_labels = top5_pie[col_groupby].values
    plt_sales = top5_pie[col_val].values

    # построение piechart c долями
    fig1, ax1 = plt.subplots()
    ax1.pie(plt_sales, labels=plt_labels, autopct='%1.1f%%', startangle=90)
    ax1.axis('equal')
    plt.title(f'Доля суммарных продаж, {group_name} ({region_name})', fontsize=13)
    plt.show()
```

Дополнительно посмотрим, как обстоят дела со **средними продажами**:

```python
def box_describe(df_name, col_val, col_groupby, region_name, group_name, figsize=(10, 5),
                 xlabel=None, ylabel='Продажи (млн долларов)', ylim=None, describe=True):
    
    """Функция возвращает боксплот и сводную статистику:
    - df_name - наименование датафрейма;
    - col_val - продажи региона;
    - col_groupby - наименование колонки для группировки;
    - region_name - наименование региона (в именительном падеже);
    - group_name - наименование группирующего признака (в именительном падеже);
    - figsize - значение figsize;
    - xlabel  - значение xlabel;
    - ylabel (по умолчанию 'Продажи (млн долларов)') - значение ylabel;
    - ylim - значение ylim;
    - describe (по умолчанию True) - нужно ли выводить сводную статистику."""
    
    # построение боксплот
    (df_name.boxplot(column=col_val, by=col_groupby, figsize=figsize))
    plt.title(f'Изучение средних продаж, {group_name} ({region_name})')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    # настройка на случай указания лимита
    plt.ylim(ylim)
    plt.show()
    
    if describe:
        # вывод сводной статистики
        df_describe = (df_name.groupby(col_groupby)[col_val].describe()
                                                             .sort_values(by='50%', ascending=False)
                      )
        return df_describe
```

## Шаг 5. Проверка гипотез

### Гипотеза 1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

**XOne** - это новая платформа, которая появилась **в 2013 году**, т. е. уже в **современный период**.  
**PC** - универсальная платформа с долгой историей, которой пользовалось несколько поколений пользователей. Поэтому будет разумно взять оценки для PC не за весь период существования платформы, а только для **современного этапа**, т. е. **после 2012 года**.

**Нулевая гипотеза H₀**: средние пользовательские рейтинги платформ Xbox One и PC **одинаковые**, т. е. разницы между оценками нет.
**Альтернативная гипотеза H₁**: средние пользовательские рейтинги платформ Xbox One и PC **не одинаковые**, т. е. оценки разные.

Сделаем **срезы с оценками пользователей** по обеим платформам. Из этих срезов возьмем Series, которые будут использованы в качестве выборочного массива для сравнения.

```python
# срез для XOne за год > 2012
sample_xone_2012 = games_region[games_region['user_score'].notnull()].query('platform == "XOne"')['user_score']

print('Количество элементов выборки:', len(sample_xone_2012))
print('Средняя пользовательская оценка: {:.2f}'.format(sample_xone_2012.mean()))

# первые несколько значений
print(sample_xone_2012.values[0:30])
```

```python
# срез для PC за год > 2012
sample_pc_2012 = games_region[games_region['user_score'].notnull()].query('platform == "PC"')['user_score']

print('Количество элементов выборки:', len(sample_pc_2012))
print('Средняя пользовательская оценка: {:.2f}'.format(sample_pc_2012.mean()))

# первые несколько значений
print(sample_pc_2012.values[0:30])
```

```python
# выведем параметры Series для XOne и PC
xone_vs_pc_2012 = pd.concat([sample_xone_2012.describe(), sample_pc_2012.describe()], axis=1)
xone_vs_pc_2012.columns = ['xone', 'pc']
xone_vs_pc_2012
```

Ранее мы уже рассматривали пользовательские оценки для **всего датафрейма**: распределение оценок было **унимодально**, но не симметрично **со скошенностью влево**. Посмотрим, как выглядит **распределение выборок** на одном графике:

```python
# на одном графике построим распределение данных
sns.distplot(sample_xone_2012, kde=True, label='Xone') # XOne
sns.distplot(sample_pc_2012, kde=True, label='PC') # PC

plt.title('Распределение выборок платформ Xone и PC')
plt.xlabel('Количество баллов (0 - минимум, 10 - максимум)')
plt.legend()
plt.grid()
plt.show()
```

Визуально оба распределения похожи: они **унимодальны** и имеют **скошенность влево**.

Для проверки **гипотезы о равенстве двух выборочных средних** используем **парный t-тест для независимых выборок**.  
Стандартным пороговым значением p-уровня значимости является значение **0.05** - установим такой порог.

```python
# установка порога p-уровня значимости
alpha = .05

# применение теста
results = st.ttest_ind(sample_xone_2012, sample_pc_2012)

# вывод результатов
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Даже при стандартном пороге p-уровня значимости у нас **не получилось отвергнуть нулевую гипотезу**, т. е. вероятность получить такие или ещё более выраженные различия оценок между платформами **случайно** не такая уж и маленькая.  
Значит, есть основания принять **нулевую гипотезу**, что средние пользовательские оценки для платформ **XOne** и **PC одинаковы**.

**Альтернативная гипотеза H₁**: средние пользовательские рейтинги жанров action и sports **не одинаковые**, т. е. оценки разные.

Также сделаем **срезы с оценками пользователей** по обоим жанрам и используем их в качестве выборочного массива для сравнения.
